```python
import numpy as np
import matplotlib.pyplot as plt

# Data arrays extracted from the table
sizes = np.array([4096, 8192, 16384])
serial_times = np.array([0.000364, 0.000877, 0.001649])
openmp_times = np.array([0.035596, 0.051807, 0.100108])
mpi_times = np.array([0.000109, 0.000071, 0.000047])

# Plotting code
plt.figure(figsize=(10, 6))
plt.plot(sizes, serial_times, 'o-', label='Serial')
plt.plot(sizes, openmp_times, 's-', label='OpenMP (4)')
plt.plot(sizes, mpi_times, 'd-', label='MPI (4)')

plt.xlabel('Problem Size (n)')
plt.ylabel('Execution Time (s)')
plt.title('Parallel vs Serial Execution Times')
plt.legend()
plt.grid(True)
plt.xscale('log')
plt.yscale('log')  # Using log scale since times vary significantly
plt.xticks(sizes, [str(size) for size in sizes])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data arrays extracted from the table
sizes = np.array([4096, 8192, 16384])
openmp_speedup = np.array([0.01, 0.02, 0.02])
mpi_speedup = np.array([1.50, 1.46, 2.79])

# Plotting code
plt.figure(figsize=(10, 6))
plt.plot(sizes, openmp_speedup, 's-', label='OpenMP (4)')
plt.plot(sizes, mpi_speedup, 'd-', label='MPI (4)')

plt.xlabel('Problem Size (n)')
plt.ylabel('Speedup')
plt.title('Speedup vs Input Size')
plt.legend()
plt.grid(True)
plt.xticks(sizes, [str(size) for size in sizes])
plt.axhline(y=1, color='gray', linestyle='--')  # Reference line for speedup = 1

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data arrays extracted from the table
threads = np.array([2, 4])
times = np.array([0.040221, 0.051807])
speedup = np.array([0.02, 0.02])
efficiency = np.array([0.01, 0.005])

# Create figure with subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('OpenMP Performance (Size=8192)')

# Time plot
ax1.plot(threads, times, 'o-')
ax1.set_xlabel('Number of Threads')
ax1.set_ylabel('Time (s)')
ax1.set_xticks(threads)
ax1.grid(True)

# Speedup plot
ax2.plot(threads, speedup, 's-')
ax2.set_xlabel('Number of Threads')
ax2.set_ylabel('Speedup')
ax2.set_xticks(threads)
ax2.axhline(y=1, color='gray', linestyle='--')
ax2.grid(True)

# Efficiency plot
ax3.plot(threads, efficiency, 'd-')
ax3.set_xlabel('Number of Threads')
ax3.set_ylabel('Efficiency')
ax3.set_xticks(threads)
ax3.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data arrays extracted from the table
processes = np.array([2, 4])
times = np.array([0.000044, 0.000047])
speedup = np.array([5.54, 2.79])
efficiency = np.array([2.77, 0.70])

# Create figure with subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('MPI Performance (Size=16384)')

# Time plot
ax1.plot(processes, times, 'o-', color='tab:red')
ax1.set_xlabel('Number of Processes')
ax1.set_ylabel('Time (s)')
ax1.set_xticks(processes)
ax1.grid(True)

# Speedup plot
ax2.plot(processes, speedup, 's-', color='tab:green')
ax2.set_xlabel('Number of Processes')
ax2.set_ylabel('Speedup')
ax2.set_xticks(processes)
ax2.axhline(y=1, color='gray', linestyle='--')
ax2.grid(True)

# Efficiency plot
ax3.plot(processes, efficiency, 'd-', color='tab:blue')
ax3.set_xlabel('Number of Processes')
ax3.set_ylabel('Efficiency')
ax3.set_xticks(processes)
ax3.axhline(y=1, color='gray', linestyle='--')  # Ideal efficiency reference
ax3.grid(True)

plt.tight_layout()
plt.show()
```
